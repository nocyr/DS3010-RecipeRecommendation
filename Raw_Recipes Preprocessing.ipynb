```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_raw_r = pd.read_csv('RAW_recipes.csv')
df_raw_r


```

```python
null_counts = df_raw_r.isnull().sum()
null_counts

```

```python
df_raw_r_cleaned = df_raw_r.dropna()
df_raw_r_cleaned
```

```python
num_rows_after_removal = df_raw_r_cleaned.shape[0]
print(num_rows_after_removal)
```

```python
import ast

# convert the 'nutrition' column from string  of lists to actual lists
df_raw_r['nutrition'] = df_raw_r['nutrition'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# defining column names for nutritional info 
nutrition_columns = ['calories', 'total_fat', 'sugar', 'sodium', 'protein', 'saturated_fat', 'fiber']

#Create a new DataFrame by splitting the 'nutrition' column into separate columns
nutrition_df = pd.DataFrame(df_raw_r['nutrition'].tolist(), columns=nutrition_columns)

#merge new nutrition
df_raw_r = pd.concat([df_raw_r.drop(columns=['nutrition']), nutrition_df], axis=1)
print(df_raw_r.head())
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_raw_r['protein'], df_raw_r['sodium'], alpha=0.5, color='green')
plt.title('Protein vs Sodium Content', fontsize=14)
plt.xlabel('Protein (g)', fontsize=12)
plt.ylabel('Sodium (mg)', fontsize=12)
plt.grid(alpha=0.7)
plt.show()
```

```python
#Create a new column 'complexity_score' by adding ingredients and steps
df_raw_r['complexity_score'] = df_raw_r['n_ingredients'] + df_raw_r['n_steps']

#Create a new timecat based on the miunutes column
df_raw_r['time_category'] = df_raw_r['minutes'].apply(lambda x: 'quick' if x <= 30 else ('medium' if x <= 60 else 'long'))

#join list of ingredients into single string 
df_raw_r['ingredients_flat'] = df_raw_r['ingredients'].apply(lambda x: ', '.join(eval(x)))
```

```python
time_category_counts = df_raw_r['time_category'].value_counts()
plt.figure(figsize=(10, 6))
time_category_counts.plot(kind='bar', color=['blue', 'green', 'red'], edgecolor='k')
plt.title('Number of Recipes by Time Category', fontsize=14)
plt.xlabel('Time Category', fontsize=12)
plt.ylabel('Number of Recipes', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns

#Calculate correlation matrix
nutrition_corr = df_raw_r[['calories', 'total_fat', 'sugar', 'sodium', 'protein', 'saturated_fat', 'fiber']].corr()
plt.figure(figsize=(10, 8))

# Create a heatmap to visualize
sns.heatmap(nutrition_corr, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Nutritional Values', fontsize=14)
plt.show()
```

```python
# count num  of recipes for each contributor and get the top 10 contributors
top_contributors = df_raw_r['contributor_id'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_contributors.plot(kind='bar', color='skyblue', edgecolor='k')
plt.title('Top 10 Contributors by Number of Recipes', fontsize=14)
plt.xlabel('Contributor ID', fontsize=12)
plt.ylabel('Number of Recipes', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
avg_nutrition_by_year = df_raw_r.groupby('submitted')[['calories', 'protein', 'sodium']].mean()
avg_nutrition_by_year.plot(kind='line', figsize=(12, 8), marker='o')
plt.title('Nutritional Trends Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Nutritional Value', fontsize=12)
plt.legend(title='Nutritional Metric')
plt.grid()
plt.show()
```
